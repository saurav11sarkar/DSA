```python
class Node:
  def __init__(self,value):
    self.value = value
    self.next = None

class LinkList:
  def __init__(self,value):
    new_node = Node(value)
    self.head = new_node
    self.tail = new_node
    self.length = 1

  def Print_list(self):
    temp = self.head
    while temp is not None:
      print(temp.value)
      temp = temp.next

  def Append_list(self,value):
    new_node = Node(value)
    if self.length == 0:
      self.head = new_node
      self.tail = new_node
    else:
      self.tail.next = new_node
      self.tail = new_node
    self.length += 1

  def Pop_list(self):
    if self.length == 0:
      return None
    temp = self.head
    pre = self.head
    while temp.next:
      pre = temp
      temp = temp.next
    self.tail = pre
    self.tail.next = None
    self.length -= 1
    if self.length == 0:
      self.head = None
      self.tail = None
    return temp

  def Fist_append_list(self,value):
    new_node = Node(value)
    if self.length == 0:
      self.head = new_node
      self.tail = new_node
    else:
      new_node.next = self.head
      self.head = new_node
    self.length += 1

  def Fist_pop_list(self):
    if self.length == 0:
      return None
    temp = self.head
    self.head = self.head.next
    self.tail = None
    self.length -= 1
    if self.length == 0:
      self.head = None
      self.tail = None
    return temp

  def Get_list(self,index):
    if index < 0 or index >= self.length:
      return None
    temp = self.head
    for _ in range(index):
      temp = temp.next
    return temp

  def Set_list(self,index,value):
    temp = self.Get_list(index)
    if temp:
      temp.value = value
      return True
    return False

  def Insert_list(self,index,value):
    if index < 0 or index > self.length:
      return False
    if index == 0:
      return self.Fist_append_list(value)
    if index == self.length:
      return self.Append_list(value)
    new_node = Node(value)
    temp = self.Get_list(index-1)
    new_node.next = temp.next
    temp.next = new_node
    self.length += 1
    return temp

  def Remove_list(self,index):
    if index < 0 or index >= self.length:
      return False
    if index == 0:
      return self.Fist_pop_list()
    # if index == self.length - 1:
    #   return self.Pop_list()
    pre = self.Get_list(index-1)
    temp = pre.next
    pre.next = temp.next
    temp.next = None
    self.length -= 1
    return temp.value

  def Reverse_list(self):
    temp = self.head
    self.head = self.tail
    self.tail = temp
    after = temp.next
    before = None
    for _ in range(self.length):
      after = temp.next
      temp.next = before
      before = temp
      temp = after








my_link_list = LinkList(2)
my_link_list.Append_list(3)
my_link_list.Append_list(4)
# my_link_list.Fist_append_list(1)
# my_link_list.Print_list()
# print(my_link_list.Pop_list().value)
# print(my_link_list.Fist_pop_list().value)
# my_link_list.Get_list(1).value
# my_link_list.Set_list(0,1)
# my_link_list.Insert_list(0,1)
# my_link_list.Insert_list(4,5)
# my_link_list.Remove_list(2)
# my_link_list.Reverse_list()
my_link_list.Print_list()
```
